```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Upload the Excel file
from google.colab import files
uploaded = files.upload()  # This will prompt you to upload a file
```

```python
# Get the file name from the uploaded dictionary
file_name = list(uploaded.keys())[0]  # Extract the uploaded file name

# Load the Excel file
df = pd.read_excel(file_name)  # Directly read the uploaded Excel file

# Display column names for reference
print("Column Names:", df.columns)
```

```python
# Ensure the correct column name for employment status
employment_column = "Employment Status"  # Change this if needed
if employment_column not in df.columns:
    print(f"Column '{employment_column}' not found! Check column names above.")
else:
    # Count unique employment statuses
    employment_counts = df[employment_column].value_counts()

    # Plot the bar chart
    plt.figure(figsize=(8, 5))
    employment_counts.plot(kind='bar', color='skyblue', edgecolor='black')

    # Customize the chart
    plt.title("Employment Status Distribution")
    plt.xlabel("Employment Status")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import os

# Get the previously uploaded file from the Colab environment
file_name = [f for f in os.listdir() if f.endswith('.xlsx')][0]  # Finds the first uploaded Excel file

# Load the Excel file
df = pd.read_excel(file_name)  # Directly read the uploaded Excel file

# Display column names for reference
print("Column Names:", df.columns)
```

```python
# Ensure the correct column name for employment status
employment_column = "Employment Status"  # Change this if needed

if employment_column not in df.columns:
    print(f"Column '{employment_column}' not found! Check column names above.")
else:
    # Define the specific responses we want to include
    valid_responses = ["Homeless", "Unemployed", "Asylum seeker", "Unable to work (please specify)", "Employed part-time"]

    # Filter the dataset to keep only these responses
    filtered_df = df[df[employment_column].isin(valid_responses)]

    # Count unique employment statuses
    employment_counts = filtered_df[employment_column].value_counts()

    # Plot the cleaned bar chart
    plt.figure(figsize=(8, 5))
    employment_counts.plot(kind='bar', color='skyblue', edgecolor='black')

    # Customize the chart
    plt.title("Filtered Employment Status Distribution")
    plt.xlabel("Employment Status")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (assuming it is already uploaded in Google Colab)
file_name = "gross-disposable-household-income-per-head-map-data.csv"

# Read the file while skipping metadata rows
df = pd.read_csv(file_name, skiprows=7)

# Sort the data by income (ascending) and select the bottom 15 areas
df_sorted_lowest = df.sort_values(by="Value (£)", ascending=True).head(15)

# Plot a bar chart for the lowest 15 areas
plt.figure(figsize=(12, 6))
plt.bar(df_sorted_lowest["Area name"], df_sorted_lowest["Value (£)"], color='red', edgecolor='black')

# Customize the chart
plt.title("Bottom 15 Areas - Gross Disposable Household Income per Head", fontsize=14)
plt.xlabel("Area Name", fontsize=12)
plt.ylabel("Income (£)", fontsize=12)
plt.xticks(rotation=45, ha="right", fontsize=10)  # Rotate labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
