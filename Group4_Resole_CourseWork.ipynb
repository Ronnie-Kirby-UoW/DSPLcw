```python
# fill all the missing value in data with the mode

Mode_nation = data['Nationality '].mode()
data['Nationality '].fillna(Mode_nation, inplace=True)
```

```python
shoereason = data['What are the shoes needed for? '].mode()
data['What are the shoes needed for? '].fillna(shoereason, inplace=True)
```

```python
shoe_size = data['Shoe Size (UK) \nPlease enter a UK shoe size. '].mode()
data['Shoe Size (UK) \nPlease enter a UK shoe size. '].fillna(shoe_size, inplace=True)
```

```python
asylum = data['Are you seeking asylum in the UK? (Refugee) '].mode()
data['Are you seeking asylum in the UK? (Refugee) '].fillna(asylum, inplace=True)
```

```python
# fill missing Age with the mean
employment = data['Employment Status'].mode()
data['Employment Status'].fillna(employment, inplace=True)
```

```python
# fill missing Sex with the mode
uc = data['Do you receive Universal Credit?'].mode()
data['Do you receive Universal Credit?'].fillna(uc, inplace=True)
```

```python
# fill missing Tumor_Size with the mean
educa = data['Are you in education or training? '].mode()
data['Are you in education or training? '].fillna(educa, inplace=True)
```

```python
# fill missing Regional_Node_Examined with the mean
lookafter = data['Are you a looked after child or care leaver? '].mode()
data['Are you a looked after child or care leaver? '].fillna(lookafter, inplace=True)
```

```python
hear = data['Where did you hear about Resole? '].mode()
data['Where did you hear about Resole? '].fillna(hear, inplace=True)
```

```python
list(data.columns)
```

```python
# perform a complete case analysis to remove all nan
data = data.dropna()
```

```python
# # construct a bar graph for the variables
T_Stage_fig = px.bar(data, x='Employment Status')
N_Stage_fig = px.bar(data, x='Do you receive Universal Credit?')
Sixth_Stage_fig = px.bar(data, x='Are you seeking asylum in the UK? (Refugee) ')
Estrogen_Status_fig = px.bar(data, x='Are you in education or training? ')

T_Stage_fig.show()
N_Stage_fig.show()
Sixth_Stage_fig.show()
Estrogen_Status_fig.show()
```
